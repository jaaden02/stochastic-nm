```python
import os

# 1. Update this path to your frames folder
folder_path = "./temp_frames"

# 2. Get all png files and sort them to ensure correct order
files = [f for f in os.listdir(folder_path) if f.endswith('.png')]
files.sort()

print(f"Found {len(files)} frames. Renumbering...")

# 3. Rename them sequentially (0, 1, 2, 3...)
for index, filename in enumerate(files):
    old_path = os.path.join(folder_path, filename)
    
    # New name: frame_000.png, frame_001.png, etc.
    new_name = f"frame_{index:03d}.png"
    new_path = os.path.join(folder_path, new_name)
    
    os.rename(old_path, new_path)

print(f"Done! Frames are now numbered 000 to {len(files)-1:03d}")
```

## 1D Bownian Motion

*Matringale:*

$dX_t = \mu(X_t,t)dt + \sigma(X_t,t) dW_t $

No drift: 

$\mu(X_t,t) = 0 $

BM properties:

$$
\begin{aligned}
&\textbf{Independent increments.}\quad \text{For } 0 \le s \le t,\; W_t - W_s \sim \mathcal{N}(0,\,t-s),\\
&\qquad \text{and } W_t - W_s \text{ is independent of } \mathcal{F}_s := \sigma\!\left(W_u:\,0 \le u \le s\right).\\[0.6em]
&\textbf{Continuous trajectories.}\quad \text{For } \mathbb{P}\text{-a.e. } \omega \in \Omega,\; t \mapsto W_t(\omega) \text{ is continuous.}
\end{aligned}
$$

```python
import numpy as np
import matplotlib.pyplot as plt


# step by step brownian motion

def bm_step():
    return np.random.normal(0,1)

def bm(steps):
    path = [0]
    for _ in range(steps):
        path.append(path[-1] + bm_step())
    return np.array(path)

def bm_vectorized(steps):
    steps_array = np.random.normal(0,1,steps)
    path = np.cumsum(np.insert(steps_array, 0, 0))
    return path

steps = 1000

t = np.linspace(0, steps, steps + 1)

# path = bm_vectorized(steps)
for i in range(200):
    plt.plot(bm_vectorized(steps))
    i += 1
plt.plot(t,np.sqrt(2*t*np.log(np.log(t))), 'k--', label='Upper P-a-s Bound')
plt.plot(t,-np.sqrt(2*t*np.log(np.log(t))), 'k--', label='Lower P-a-s Bound')
plt.title("Brownian Motion Path")
plt.xlabel("Steps")
plt.ylabel("Position")
plt.legend()
plt.show()
```
